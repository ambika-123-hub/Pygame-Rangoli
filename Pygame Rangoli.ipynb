```python
import pygame
from pygame.locals import *
# initiate pygame and give permission to use pygame's functionality
pygame.init()

# create the display surface object (screen) of specific dimension
screen=pygame.display.set_mode((600,600))
# set caption
pygame.display.set_caption("Basic Shapes")

#fill the screen with black color
screen.fill((0,0,0))


def drawingfunction(x,y,width,height):
    #using draw.rect module of pygame, draw the outlined rectangle
    # Syntax: pygame.rect(surface,color,rect,width)
    for i in range(51):
        pygame.draw.rect(screen,(0,25+i*4,25+i*4),[x-i,y-i,width+2*i,height+2*i],5)

    # using pyagame.draw.circle(surface,color,center,radius,width) we can draw circle
    
    circle_x = width/2 + x
    circle_y = height/2 + y
    if height<width:
        radius=height/2
    else:
        radius=width/2
        
    for i in range(51):
        
        pygame.draw.circle(screen,(25+i*4,i,0),[circle_x, circle_y],radius*0.02*i,15)
    
    for t in range(0,20,2):
        for i in range(20):
            pygame.draw.circle(screen,(255,25+i*10,0),[circle_x, circle_y/2],5*i,5)
            pygame.draw.circle(screen,(255,25+i*10,0),[circle_x, 1.5*circle_y],5*i,5)
            pygame.draw.circle(screen,(255,25+i*10,0),[circle_x/2, circle_y],5*i,5)
            pygame.draw.circle(screen,(255,25+i*10,0),[circle_x*1.5, circle_y],5*i,5)

            pygame.draw.circle(screen,(255,25+i*10,0),[circle_x*0.675, circle_y*0.63],5*i,5)
            pygame.draw.circle(screen,(255,25+i*10,0),[circle_x*1.35, circle_y*1.35],5*i,5)
            pygame.draw.circle(screen,(255,25+i*10,0),[circle_x*1.38, circle_y*0.67],5*i,5)
            pygame.draw.circle(screen,(255,25+i*10,0),[circle_x*0.6, circle_y*1.31],5*i,5)

            pygame.draw.circle(screen,(10*i,51+i*2,0),[circle_x, circle_y],radius*0.02*i,5)
        
    
drawingfunction(50,50,500,500)
#draw the surface object to the screen
pygame.display.update()

#------------------------------------------------------
status = True
while (status):
    # iterate over the list of Event objects that was returned by pygame.event.get() method.
    for i in pygame.event.get():
        # if event object type is QUIT then quitting the pygame and program both.
        if i.type == pygame.QUIT:
            status = False

# deactivates the pygame library
pygame.quit()
```
